## Introduction:
The objective of this project is to find the best place geographically to place a ghost kitchen in Texas. Ghost kitchens are small kitchens that are only avalable to order online through 3rd party logistics such as uber, doordash, and grubhub. These ghost kitchens leverage low start up costs and reduce risk by running multiple brands to achieve profability in their business model. 


Being able to provide an idea of not only the medium house income of a targeted area but also know if the people in these areas are searching for food would be very valuable information for ghost kitchen investors. The areas with higher food searches and high medium home income wouuld help them have higher potential for success.
